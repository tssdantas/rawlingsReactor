```python
import plotly.express as px
import pandas as pd
```

```python
data = []
with open('output.csv','r') as f:
    #lines = f.readlines()
    while line := f.readline():
        currentline = line.rstrip().split(",")
        data.append(currentline[:5])
```

```python
df = pd.DataFrame(data, columns=['x', 'y1', 'y2', 'y3', 'y4'])
```

```python
df.head()
```

```python

fig = px.line(df, x='x', y=['y1', 'y2', 'y3', 'y4'])
fig.update_yaxes(autorange="reversed")
fig.show()
```
